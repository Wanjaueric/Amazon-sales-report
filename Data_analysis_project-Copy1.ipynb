# Data Analysis

# 1.	Which regions have the highest total sales revenue?

```python
Highest_total_revenue = df.groupby(df['Region'])['Total Revenue'].sum()
Highest_total_revenue.idxmax()
```

```python
group_data = df.groupby(df['Region'])['Total Revenue'].sum()

sns.set_style( 'darkgrid' )
plt.figure(figsize = (15,10))
sns.lineplot(data = group_data, linestyle = '--', marker = 'o', markersize = 10, markerfacecolor = 'red', markeredgecolor = 'black')

plt.title("Highest Total Sales Revenue by Region", color = 'Blue', fontsize = 20)
plt.xlabel('Regions', color = 'Red', fontsize = 20)
plt.ylabel('Total Revenue', color = 'Red', fontsize = 20)
plt.show()
```

# 2.	What is the average unit price and unit cost for each item type?

```python
Avg_unit_price = df.groupby(df['Item Type'])['Unit Price'].mean()
Avg_unit_cost = df.groupby(df['Item Type'])['Unit Cost'].mean()

Avg_price_cost = pd.DataFrame({'Average unit price':Avg_unit_price,
                             'Average unit cost': Avg_unit_cost})
Avg_price_cost
```

# 3.Which country has the highest total profit?

```python
Total_profit_by_country = df.groupby(df['Country'])['Total Profit'].sum()
Highest_Total_profit_by_country  = Total_profit_by_country.idxmax()
print('The Country with the highest profit is', Highest_Total_profit_by_country)
```

```python
group_data = df.groupby(df['Country'])['Total Profit'].sum()

sns.set_style( 'darkgrid' )
plt.figure(figsize = (15,10))
sns.barplot(x = group_data.index, y = group_data)

plt.title('Country with Highest total profit', fontsize = 20, color = 'Blue')
plt.xlabel('Country', fontsize = 20, color = 'Red')
plt.xticks(rotation = 90)
plt.ylabel('Total Profit', fontsize = 20, color = 'Red')
plt.show()
```

# 4.	How does the sales channel affect the order priority distribution?

```python
sales_channel = df.groupby(df['Sales Channel'])['Order Priority'].value_counts()
sales_channel
```

```python
sales_channel_order_priority_distribution= df.groupby(df['Sales Channel'])['Order Priority'].count()

# reset the indexto convert the grouped data into a Dataframe
sales_channel_order_priority_distribution = sales_channel.reset_index(name = 'Count')

# set the style
#plt.style.use('fivethirtyeight')
sns.set_style('darkgrid')
plt.figure(figsize = (15,10))
sns.barplot(data =sales_channel_order_priority_distribution , x = 'Sales Channel', y = 'Count', hue = 'Order Priority')

#add labels
plt.title('Sales channel order priority distribution')
plt.xlabel(' Sales Channel')
plt.ylabel('Count')
plt.show()
```

# 5.	What is the average order processing time (duration between order and ship dates) for each sales channel?

```python
df['Processing Time'] = df['Ship Date'] - df['Order Date']
Avg_Processing_Time = df.groupby(df['Sales Channel'])['Processing Time'].mean()
Avg_Processing_Time
```

# 6.Which items types have the highest and lowest total sales?

```python
group_item_type = df.groupby(df['Item Type'])['Total Revenue'].sum()
highest_sales_revenue_item_type = group_item_type.idxmax()
lowest_sales_revenue_item_type = group_item_type.idxmin()

print('Highest sales Revenue by Item Type', highest_sales_revenue_item_type)
print('Lowest sales Revenue by Item Type',lowest_sales_revenue_item_type)
```

```python
plt.figure(figsize =(10,5))
plt.style.use('seaborn-v0_8-darkgrid')

# highest max value
sns.scatterplot(x = group_item_type.index, y = group_item_type, s = 200)
max_index = group_item_type.idxmax()
plt.scatter(x = max_index, y = group_item_type[max_index], s = 200, color = 'Green', edgecolor = 'Black')

#lowest max value
min_index = group_item_type.idxmin()
plt.scatter(x = min_index, y = group_item_type[min_index], s = 200, color = 'Red', edgecolor = 'Black')

plt.title('Scatter plot showing highest and lowest sales per Item Type')
plt.xticks(rotation = 60)

plt.show()
```

# 7.	How does the order priority vary across different regions?

```python
Regions_by_order_priority = df.groupby(df['Region'])['Order Priority'].value_counts().reset_index(name = 'Order_priority_Count')
Regions_by_order_priority
```

```python
Regions_by_order_priority = df.groupby(df['Region'])['Order Priority'].value_counts().reset_index(name = 'Order Priority Count')

plt.figure(figsize = (15,10))

sns.barplot(data = Regions_by_order_priority, x= 'Region', y = 'Order Priority Count' , hue = 'Order Priority')

plt.title('Order of priority per Regions')
plt.xticks(rotation = 90)
plt.yticks(np.arange(1,15,1))

plt.show()
```

# 8.	What is the correlation between unit price and total profit?

```python
correlation_unit_price_total_profit = df['Unit Price'].corr(df['Total Profit'])
print('Correlation between Unit price and Total Profit:' ,correlation_unit_price_total_profit)
```

```python
plt.figure(figsize = (10,5))
plt.scatter(x = correlation_unit_price_total_profit, y = correlation_unit_price_total_profit, s = 200, color = 'Red')

plt.title('Correlation_Unit_price_Total_profit')
plt.xticks(np.arange(-1, 2, 0.5))
plt.yticks(np.arange(-1, 2, 0.5))

plt.show()
```

# 9.	Are there any seasonal trends or patterns in the sales data?

```python
month_names= {1: 'JAN',
              2: 'FEB',
              3: 'MAR',
              4: 'APR',
              5: 'MAY',
              6: 'JUN',
              7: 'JUL',
              8: 'AUG',
              9: 'SEPT',
             10: 'OCT',
             11: 'NOV',
             12: 'DEC'}
monthly_sales = df.groupby(df['Order Date'].dt.month)['Total Revenue'].sum()
monthly_sales.index= monthly_sales.index.map(month_names)

monthly_sales
```

```python
sns.barplot(x= monthly_sales.index, y= monthly_sales)
plt.title('Month Wise Total Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```
